```python
# Fetch data using ergast API

year = 2023  # Specify the year you want to fetch data for
url = (f"http://ergast.com/api/f1/{year}/results.json?limit=1000")
response = requests.get(url)
data = response.json()
```

```python
# Instead of ergast, let's use the Kaggle csv
# Source: https://www.kaggle.com/datasets/rohanrao/formula-1-world-championship-1950-2020

# The csv files are all in ./archive

# Import every csv file in the archive directory into an individual DataFrame
import os
archive_path = './archive'
dfs = {}
for filename in os.listdir(archive_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(archive_path, filename)
        df_name = filename.replace('.csv', '')
        dfs[df_name] = pd.read_csv(file_path)

# Print the names of the DataFrames
print('Datframes:')
for df_name in dfs.keys():
    print(df_name)

```

```python

# Extract relevant data from the JSON structure
races = data['MRData']['RaceTable']['Races']
# Convert to DataFrame (you may need to flatten nested structures)
# Each race contains a 'Results' list; flatten all results across races
results = []
for race in races:
    for result in race['Results']:
        # Add race info to each result
        result['raceName'] = race['raceName']
        result['round'] = race['round']
        result['date'] = race['date']
        result['circuitName'] = race['Circuit']['circuitName']
        results.append(result)
# Now normalize the results list
df = pd.json_normalize(results)
df.head()
```
